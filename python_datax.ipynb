```python
# Removed row with two NULLs
df = df.drop(26)

# Added a gender to the row missing a gender
df.loc[8,"gender"] = "male"
```

```python
df['class'] = df['class'].replace('Fifth', 'Five')
```

```python
# Filter students who achieved the average or above
df[df["mark"] >= df["mark"].mean()]
```

```python
# Sort students by mark from highest to lowest
df = df.sort_values(by="mark", ascending=False)
df.head()
```

```python
# Creates a new column called grade and assigns a grade based on each student's mark

df.loc[df["mark"] >= 85, "grade"] = "A"
df.loc[(df["mark"] >= 70) & (df["mark"] <= 84), "grade"] = "B"
df.loc[(df["mark"] >= 60) & (df["mark"] <= 69), "grade"] = "C"
df.loc[df["mark"] < 60, "grade"] = "D"

df.head()
```
