```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student.csv')
```

```python
# Removed row with two Nulls
df = df.drop(26)

# Added a gender to the row missing a gender
df.loc[8,"gender"] = "male"

```

```python
plt.figure(figsize=(10,6))
plt.hist(df['mark'], bins=10, edgecolor='black', color='yellow')
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=df, color='#F3CFC6')
plt.title('Number of Students in Each Class')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Replace the value 'Fifth' with 'five' in the class column
df['class'] = df['class'].replace('Fifth', 'Five')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='mark', data=df, hue='class', palette='Set3', legend=False)
plt.title('Marks Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Marks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=df, color='#F3CFC6')
plt.title('Number of Students in Each Class')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='gender', y='mark',data=df, hue='gender',  palette={'female': '#DE3163', 'male': 'skyblue'}, alpha=0.7)
plt.title('Marks by Gender')
plt.xlabel('Gender')
plt.ylabel('Marks')
plt.show()
```

```python
# Calculates average mark per class
grouped = df.groupby("class")["mark"].mean()

plt.figure(figsize=(8, 6))
plt.bar(grouped.index, grouped.values, color='skyblue')

plt.title('Average Marks by Class')
plt.xlabel('Class')
plt.ylabel('Average Mark')

plt.show()
```
